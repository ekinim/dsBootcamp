### Soru 1:
Bu dersimizde görmediğimiz fakat çok yaygın olarak kullanılan çizge türlerinden birisi de çubuk grafikleri (bar plot)'dir. Çubuk grafikleri özellikle, kategorik (yani sınırlı sayıda farklı değer içeren) değişkenleri incelemek için yararlı araçlardır. Bu ödevde, kategorik bir değişken içeren bir veri kümesi bulmanız gerekecek. Bunun için Kaggle sitesindeki veri setlerine bir göz atabilirsiniz. Daha sonra, bulduğunuz veri kümesini Pandas'ı kullanarak bir veri çerçevesine aktarın. Veri kümenizde yer alan kategorik değişkenlerden iki tanesini seçin ve subplot kullanarak, bu iki değişkenin çubuk grafiklerini çıkarın.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
covid19 = pd.read_csv("C:/Users/gulme/Desktop/py/covid19tr.csv")
covid19.index.name = "Index"
covid19
```

```python
plt.figure(figsize=(20,5))

plt.bar(covid19["Last_Update"], covid19["Confirmed"],
        label="Pozitif Vaka", width=0.7, align="center", color="orange", log=True, alpha=0.6)

plt.bar(covid19["Last_Update"], covid19["Recovered"],
        label="İyileşen", width=-0.35, align="edge", color="green", log=True)

plt.bar(covid19["Last_Update"], covid19["Deaths"],
        label="Ölümler", width=0.35, align="edge", color="black", log=True)

plt.title("Covid19 İstatistikleri (Türkiye)", fontsize=16)
plt.ylabel("Vaka Sayısı (log)", fontsize=18)
plt.legend(loc="upper left", fontsize=16, frameon=False)
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=16)
plt.grid(which="major", axis="y", alpha=0.4)
plt.show()
```
