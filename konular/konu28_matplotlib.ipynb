```python
np.random.seed(1221)
# Random sayıların her çalıştırmada değişmemesi, tutarlı olarak üretilmeleri için .seed() komutunu kullanıyoruz.
```

```python
df1 = pd.DataFrame()
df1["rand"] = np.random.rand(100)    # 100 adet random number
df1["rand_sq"] = df1["rand"]**2
df1["rand_shift"] = df1["rand"]+2

df1["count_sq"] = df1.index**2
df1["count_sqrt"] = np.sqrt(df1.index)
```

```python
df1
```

```python
plt.plot(df1["rand"])
plt.show()
```

## gösterim seçenekleri

```python
plt.plot(df1["rand"], color="purple")
plt.plot(df1["rand_shift"], color="green")    # ikinci bir grafik ekledik
plt.ylim([-0.1, 3.1])
plt.ylabel("Değerler")
plt.xlabel("Index")
plt.title("Random Değerler")
plt.show()
```

## scatter graph (serpme/saçım grafiği)

```python
plt.scatter(
    x=df1["rand"],
    y=df1["rand_sq"],
    color="purple",
    marker="x", s=10
)
plt.scatter(
    x=df1["rand"],
    y=df1["rand_shift"],
    color="green",
    marker="*", s=10
)
# s=10 boyutu belirtiyor.
plt.show()
```

##  Pandas'ın `plot()` fonksiyonu
"Bazı basit grafikler için Pandas'ın içinde yer alan matplotlib işlevlerinin bazılarını kullanarak da hızlıca ilerleyebiliriz. Bunların kullanımı kolaydır, ancak yeterince güçlü değildirler. Bu yüzden her durum için yeterli olmaz."

## Birden fazla grafik çizdirmek

```python
plt.figure(figsize=(20,5))    # https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.figure.html

plt.subplot(1,2,1)    # İlk iki parametre figürün boyutlarını tanımlar, üçüncü ise o noktada oluşturduğunuz subplotu tanımlar.
plt.plot(df1["rand"], color="purple")
plt.ylabel("Değerler")
plt.title("Rastgele Değerler")

plt.subplot(1,2,2)    # https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.subplot.html#matplotlib.pyplot.subplot
plt.plot(df1["rand_shift"], color="green")
plt.ylabel("Ötelenmiş Değerler")
plt.title("Ötelenmiş Diziler")
plt.show()
## .subplot()'ın kullanımını anladım fakat 1,2,1 olan argümanlarından ilk ikisini anlamadım.
# solda rastgele değerlere sahip olan dizi ve sağda bunun ötelenmiş haliyle
# 1 yüksekliğinde ve 2 genişliğinde bir subplot oluşturduk. 
```

"Evet, plt.figure() fonksiyonu, oluşturmak istediğimiz şeklin boyutunu, istediğimiz x ve y boyutlarıyla bir ikili (tuple) alarak ayarlar. Sonra her çizge plt.subplot() fonksiyonu kullanılarak oluşturulur. Bu, üzerinde çalışacağımız çizgeleri belirtir. Diğer bir deyişle, plt.subplot() fonksiyonu ilk iki parametreyle tüm figürün kaç satır ve sütundan oluştuğunu, üçüncü parametre ile de o anki çizgenin bu figürün kaçıncı çizgesi olacağını belirler."

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.scatter(y=df1["rand"], x=df1.index, color="teal")
plt.ylabel("Değerler")
plt.title("Rastgele Diziler")

plt.subplot(1,2,2)
plt.scatter(x=df1["rand_sq"], y=df1["rand"], color="red")
plt.ylabel("Kare Değerler")
plt.title("Kare Diziler")

plt.tight_layout()    # grafikleri daha kompakt oluyor, yaklaştırıyor.
plt.show()
```

# Histogram
"Histogram, bir değişkenin olası değerlerini ve bu değerlerin ne kadar yaygın olduğunu göstermek için harika bir araçtır."

```python
x1 = np.random.normal(10, 5, 1000)
plt.hist(x1)
plt.show()
# fonksiyona verdiğimiz 10 ve 5 sayıları, kullandığımız normal dağılımın merkezi ve standart sapmasını belirler.
```

## Kutu sayımı ve yerleştirme

```python
x2 = np.random.normal(0,2,1000)
plt.hist(x2, bins=50, color="red")    # bins, yani kutulardan 50 adet belirledik.
plt.xlabel("Rastgele Değerler")
plt.title("Kutu Seçimi")
plt.show()
```

"Yukarıdaki hücreyi yeniden çalıştırırsanız, temel verilerin değerlerine bağlı olarak x-ekseni ve kutu konumunun her seferinde hafifçe değişeceğini fark edersiniz. Kutuların kenarları muhtemelen yatay eksendeki tamsayı işaretleriyle tam olarak uymaz. Bunun nedeni matplotlib'e her bir kutunun olmasını istediğimiz yer yerine kaç tane kutu istediğimizi söylememizdir."

```python
x3 = np.random.normal(10, 5, 1000)
y3 = np.random.normal(15, 5, 10000)

plt.hist(x3, bins=np.arange(-10,40), color="blue", alpha=0.5)
plt.hist(y3, bins=np.arange(-10,40), color="red", alpha=0.5)
plt.title("Manuel Kutu Yerleşimi")
plt.xlabel("Rastgele Değerler")
plt.ylabel("Değer Miktarı")
plt.show()
```

## Histogramların normalleştirilmesi¶

```python
plt.hist(x3, density=True, bins=np.arange(-10,40), color="blue", alpha=0.5)
plt.hist(y3, density=True, bins=np.arange(-10,40), color="red", alpha=0.5)
plt.title("Normalize Edilmiş Histogramlar")
plt.xlabel("Rastgele Değerler")
plt.show()

## normed=True 3.2.0 versiyonunda kaldırılmış sanırım. 3.1.1'de var. density=True ile yaptım.
## https://matplotlib.org/3.2.0/api/_as_gen/matplotlib.pyplot.hist.html
```

# Boxplot (kutu) grafiği

```python
np.random.seed(111)    # .seed() ne işe yarıyor?

x4 = np.random.normal(10,5,1000)
plt.boxplot(x4)
plt.show()
```

Boxplot, **çeyrekler açıklığı** ***(interquartile range ya da kısaca IQR)*** denilen şeyi kapsayan bir kutudan oluşur. İlk olarak çeyrekler açıklığının ne olduğunu açıklayalım. Bir değişkeni, içerdiği değerlere göre aşağıdaki gibi dört çeyrek bölüme ayırabilirsiniz:

1. En düşük değerden %25’e,
2. %25’den medyana,
3. Medyandan %75’e ve
4. %75’den maksimum değere.

Bu gruplara çeyrekler (quartiles) denir ve ortadaki ikisinin birlikte kapsadığı değerlere çeyrekler arası (interquartile range) denilir. Böylece boxplot'ta gördüğümüz kutu, %25’den %75’e kadar tüm verileri içermekte ve medyanı kırmızı bir çizgi ile göstermektedir.

Kutunun üstünde ve altında bıyıklar (whiskers) denilen dikey çizgiler bulunur (bu sebepten, boxplota bazen "kutu ve bıyık çizgeleri" de denilir). Bıyıklar, IQR kutusunun üstünden ve altından 1.5x uzaklığa kadar aşağı ve yukarı uzanacaktır. Tüm değerler bu aralıkta ise, bıyıklar değişkenin maksimum ve minimum değerlerine kadar gidecektir. IQR'ın 1.5 katının ötesindeki herhangi bir değer, bıyıkların sonunun ötesinde küçük bir kene (tick) olarak temsil edilecektir. Örneğin, yukarıdaki çizgede, bıyıkların üstünde ve altında birkaç değer var.

Boxplot, bir değişkenin yoğunluğu veya dağılımı hakkında bilgi edinmek için yararlı değildir. Bununla birlikte, değer aralığını araştırmak ve olası aykırı değerleri (outlier) tespit etmek için çok yararlıdır. Bir kutu çizgesi üzerinde bir nokta ya da daire ile işaretlenmiş veri noktaları genellikle aykırı değerler olarak kabul edilir. Bu nedenle, yeni bir veri üzerinde çalışırken kutu çizgeleri yaygın olarak kullanılır. Boxplot hakkında daha fazla bilgi edinmek için .boxplot() dökümantasyonuna bakabilirsiniz.
